```python
from textblob import TextBlob
import pandas as pd
```

```python
def analyze_sentiment(text):
    """
    Analyze sentiment of user journal text using TextBlob.
    Returns sentiment label and polarity score.
    """
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity  # Range: -1 to +1

    if polarity > 0.1:
        sentiment = "Positive"
    elif polarity < -0.1:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"

    return sentiment, polarity
```

```python
def analyze_multiple_entries(entries):
    results = []

    for entry in entries:
        sentiment, score = analyze_sentiment(entry)
        results.append({
            "Journal Entry": entry,
            "Sentiment": sentiment,
            "Polarity Score": round(score, 3)
        })

    return pd.DataFrame(results)
```

```python
#example usage:
sample_entries = [
    "I feel overwhelmed with work and anxious all the time.",
    "Today was a really good day! I enjoyed spending time with friends.",
    "Nothing special happened today. Just a normal, boring day."
]

df = analyze_multiple_entries(sample_entries)
df
```
